```python
import random
import pandas as pd

# Sample employee names (you can expand this list)
employee_names = ["John Smith", "Emily Johnson", "Michael Brown", "Sarah Davis", "David Wilson", "Jessica Lee", "James Harris"]

# Sample problems and solutions
problems = [
    "Unable to access shared company drive",
    "Slow response times for customer emails",
    "Printer not connecting to the network",
    "Missing important files from project folder",
    "Confusion with new company software",
    "Cannot log into company CRM system",
    "Emails not syncing across devices",
    "Computer running slow",
    "Missed deadline due to lack of resources",
    "Difficulty accessing shared files",
    "Team communication is unclear",
    "Customer service queries piling up",
    "Employee absence without notice",
    "Employee burnout from overwork",
    "Manual data entry errors",
    "Lack of collaboration on shared projects"
]

solutions = [
    "Restart the system and check network settings",
    "Implement an email template for quick replies",
    "Restart printer and check network settings",
    "Check backup and retrieve from last sync",
    "Schedule a training session for all employees",
    "Reset the password and check internet connectivity",
    "Reconfigure email settings and ensure synchronization is enabled",
    "Clear cache and run a disk cleanup utility",
    "Request additional resources from the project manager and adjust the timeline",
    "Check permissions on the shared folder and ensure everyone has the correct access",
    "Implement a clear communication plan using project management tools",
    "Assign more staff to handle customer service during peak hours",
    "Set up a reminder system for employees to notify HR of absences in advance",
    "Monitor workloads and encourage regular breaks and time-off",
    "Automate data entry using a workflow management tool",
    "Schedule regular check-ins and use collaboration tools like Slack or Microsoft Teams"
]

# Create a list to hold the rows of data
data = []

# Generate 10,000 entries
for _ in range(10000):
    name = random.choice(employee_names)
    problem = random.choice(problems)
    solution = random.choice(solutions)
    data.append([name, problem, solution])

# Create a DataFrame and export to CSV
df = pd.DataFrame(data, columns=["employeeName", "problem", "proposedSolution"])
df.to_csv("employee_issues.csv", index=False)
```
